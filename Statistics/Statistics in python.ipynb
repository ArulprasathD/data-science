```python
import statistics  as stats

arr = [1,2,3,4,5]

print('mean',stats.mean(arr))
```

```python
print('median',stats.median(arr))
```

```python
print('mode',stats.mode(arr))
```

```python
print('Standard deviation :',stats.stdev(arr))
```

```python
print('Varience :',stats.variance(arr))
```

```python
import numpy as np

arr = [1,2,3,4,5,6,7,8]

print('mean :', np.mean(arr))
print('median : ',np.median(arr))
print('standard deviation : ',np.std(arr))
print('varience :',np.var(arr))
print('percentile : ',np.percentile(arr,30))
```
