```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load built-in dataset
df = sns.load_dataset("tips")
```

```python
sns.kdeplot(df['total_bill'], fill=True, color='green')
plt.title("KDE Plot - Total Bill")
plt.show()
```

```python
sns.displot(df['total_bill'], kde=True, color='orange')
plt.title("Displot - Total Bill")
plt.show()
```

```python
sns.ecdfplot(df['total_bill'])
plt.title("ECDF Plot - Total Bill")
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex',data=df)
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Line Plot - Size vs Total Bill")
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', kind='scatter', data=df)
plt.title("Relplot - Scatter Example")
plt.show()
```

```python
sns.barplot(x='sex', y='tip', data=df)
plt.title("Bar Plot - Tip by Sex")
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.title("Count Plot - Number of Customers per Day")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot - Total Bill by Day")
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=df, jitter=True)
plt.title("Strip Plot - Tip by Day")
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=df)
plt.title("Swarm Plot - Tip by Day")
plt.show()
```

```python
sns.boxenplot(x='day', y='total_bill', data=df)
plt.title("Boxen Plot - Total Bill by Day")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.title("Regression Plot - Total Bill vs Tip")
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title("LM Plot - Regression by Gender")
plt.show()
```

```python
sns.pairplot(df, hue='sex')
plt.title("Pair Plot - All Numeric Columns")
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Styled Plot Example - Whitegrid")
plt.show()
```
