```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
print(df.head())
print(df.columns)
type(df)
```

# histogram

```python
sns.histplot(df["tip"],kde=False,bins = 20,color = 'r')
plt.title('Histogram - tip_table')
plt.show()
```

#  KDE plot

```python
sns.kdeplot(df['tip'],fill = True,color = 'green')
plt.title('KDE-plot - tips')
plt.show()
```

# DISK plot hist + kde

```python
sns.displot(df['tip'],kde = True,color = 'green')
plt.title('Displot - tips')
plt.show()
```

```python
sns.ecdfplot(df['total_bill'])
plt.title('ECDF plot - total bill')
plt.show()
```

# relational plot

# scatter plot


```python
sns.scatterplot(x='total_bill',y='tip',hug = "sex",data=df)
plt.title('scatter plot - bill & tips')
plt.show()
```
