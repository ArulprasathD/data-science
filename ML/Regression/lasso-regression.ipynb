```python
from sklearn.linear_model import Lasso
```

```python
X = [[1], [2], [3], [4], [5]]
y = [10, 20, 30, 40, 50]
```

```python
lso= Lasso(alpha = 0.1)
```

```python
lso.fit(X,y);
```

```python
y_pred = lso.predict([[6]])
```

```python
print("prediction : ",y_pred[0])
```

```python
print(lso.coef_)
```

```python
print(lso.intercept_)
```
