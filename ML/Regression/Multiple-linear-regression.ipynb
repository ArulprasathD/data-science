```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([
    [1, 22],
    [2, 25],
    [3, 27],
    [4, 30],
    [5, 35]
])  # experience, age

y = np.array([30000, 35000, 40000, 45000, 50000])

model = LinearRegression()
model.fit(X, y)

print("Prediction (6 yrs exp, age 40):", model.predict([[6, 40]]))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = [
    [2, 25],   # Experience, Age
    [3, 28],
    [5, 32],
    [7, 35]
]

y = [30000, 40000, 55000, 70000]

model = LinearRegression()

model.fit(X,y);

y_pred = model.predict([[6,31]])

print(f"prediction of 6 yrs,31 age :{y_pred[0]:.2f}")
```
