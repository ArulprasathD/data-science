```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([
    [1, 22],
    [2, 25],
    [3, 27],
    [4, 30],
    [5, 35]
])  # experience, age

y = np.array([30000, 35000, 40000, 45000, 50000])

model = LinearRegression()
model.fit(X, y)

print("Prediction (6 yrs exp, age 40):", model.predict([[6, 40]]))
```
